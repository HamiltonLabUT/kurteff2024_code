```python
import csv
import numpy as np
from glob import glob
```

```python
# get fnames
git_path = '/home/kfsh/git/onsetProd/'
fnames = glob(git_path + 'ecog/eventfiles/S0010/S0010_B1/*.txt')
print(fnames)
```

```python
# subtract offsets
offset = 73.9
for f in fnames:
    print(f)
    adjusted_f = []
    with open(f) as tsvfile:
        reader = csv.reader(tsvfile, delimiter='\t')
        # Go through the rows
        for row in reader:
            if 'click' in f:
                print(row)
                tmin = float(row[0]) - offset
                tmax = float(row[2]) - offset
                n = float(row[4])
                adjusted_f.append([tmin, '', tmax, '', n, ''])                
            else:
                print(row)
                tmin = float(row[0]) - offset
                tmax = float(row[1]) - offset
                n = int(row[2])
                s = row[3]
                adjusted_f.append([tmin,tmax,n,s])
    # Save the file
    with open(f,'w+') as my_csv:
        csvWriter = csv.writer(my_csv,delimiter='\t')
        csvWriter.writerows(adjusted_f)
            
```
